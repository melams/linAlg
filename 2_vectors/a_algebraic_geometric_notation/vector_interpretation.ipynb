```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Algebraic & Geometric Interpretation

```python
# 2-dimensional vector
vec_2d = [3, -2 ]

vec_3d = [4, -3, 2]
# Tranpose - row to column

vec_3d_t = np.transpose(vec_3d)

print("this is a 2d row vector: ", str(vec_2d))

print("this is a 3d row vector: ", str(vec_3d))

print(vec_3d_t)
```

# Geometric Interpretation

```python
# plot them
plt.plot([0,vec_2d[0]],[0,vec_2d[1]])
plt.axis('equal')
plt.plot([-4, 4],[0, 0],'k--')
plt.plot([0, 0],[-4, 4],'k--')
plt.grid()
plt.axis((-4, 4, -4, 4))
plt.show()


# plot the 3D vector
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, vec_3d[0]],[0, vec_3d[1]],[0, vec_3d[2]])

# make the plot look nicer
plt.axis('equal')
ax.plot([0, 0],[0, 0],[-4, 4],'k--')
ax.plot([0, 0],[-4, 4],[0, 0],'k--')
ax.plot([-4, 4],[0, 0],[0, 0],'k--')
plt.show()
```
